### <ins>Problem 1</ins>: Willy Wonka Chocolate Pipeline Distribution

<i>1. Problem Solved in Excel </i> 

<i>2. Problem Solved with NetworkX package </i> 

```python
import networkx as nx
#Initializing a graph 
graph = nx.DiGraph()

#Adding all the 8 nodes
for i in range(1,9):
    graph.add_node(i)

#Adding the edges 
graph.add_edge(1,2, capacity = 20)
graph.add_edge(1,3, capacity = 15)
graph.add_edge(2,5, capacity = 15)
graph.add_edge(2,4, capacity = 10)
graph.add_edge(3,4, capacity = 13)
graph.add_edge(3,6, capacity = 15)
graph.add_edge(3,7, capacity = 10)
graph.add_edge(4,5, capacity = 10)
graph.add_edge(4,3, capacity = 13)
graph.add_edge(4,7, capacity = 12)
graph.add_edge(5,2, capacity = 15)
graph.add_edge(5,6, capacity = 7)
graph.add_edge(5,8, capacity = 10)
graph.add_edge(6,7, capacity = 8)
graph.add_edge(6,5, capacity = 7)
graph.add_edge(6,8, capacity = 8)
graph.add_edge(7,6, capacity = 8)
graph.add_edge(7,8, capacity = 10)
```

```python
flow_value, flow_dict = nx.maximum_flow(graph, _s= 1, _t = 8, capacity= "capacity")
print("The maximum chocolate flow tranported for Wonka going from node '1' to '8' is: {0} Units/hour\n".format(flow_value))
print("The path followed to reach this maximum is: {0}".format(flow_dict))
```

<i>3. Sensitivity Analysis </i> 

```python
#Increasing edges leaving from node 1 and edges coming in to node 8 by constant (k= 1.5)

k= 1.5
graph.add_edge(1,2, capacity = 20*k)
graph.add_edge(1,3, capacity = 15*k)
graph.add_edge(5,8, capacity = 10*k)
graph.add_edge(6,8, capacity = 8*k)
graph.add_edge(7,8, capacity = 10*k)
flow_value1, flow_dict1 = nx.maximum_flow(graph, _s= 1, _t = 8, capacity= "capacity")
print("The new maximum after multiplying the arc capacities by k=1.5 for Wonka going from node '1' to '8' is: {0} Units/hour\n".format(flow_value1))
print("The path followed to reach this maximum is: {0}".format(flow_dict1))
```

```python
#Increasing edges leaving from node 1 and edges coming in to node 8 by constant (k= 2.47)

k= 2.47
graph.add_edge(1,2, capacity = 20*k)
graph.add_edge(1,3, capacity = 15*k)
graph.add_edge(5,8, capacity = 10*k)
graph.add_edge(6,8, capacity = 8*k)
graph.add_edge(7,8, capacity = 10*k)
flow_value2, flow_dict2 = nx.maximum_flow(graph, _s= 1, _t = 8, capacity= "capacity")
print("The new maximum after multiplying the arc capacities by k=2 for Wonka going from node '1' to '8' is: {0} Units/hour\n".format(flow_value2))
print("The path followed to reach this maximum is: {0}".format(flow_dict2))
```

```python
#Increasing edges leaving from node 1 and edges coming in to node 8 by constant (k= 3)

k= 3
graph.add_edge(1,2, capacity = 20*k)
graph.add_edge(1,3, capacity = 15*k)
graph.add_edge(5,8, capacity = 10*k)
graph.add_edge(6,8, capacity = 8*k)
graph.add_edge(7,8, capacity = 10*k)
flow_value3, flow_dict3 = nx.maximum_flow(graph, _s= 1, _t = 8, capacity= "capacity")
print("The new maximum after multiplying the arc capacities by k=2 for Wonka going from node '1' to '8' is: {0} Units/hour\n".format(flow_value3))
print("The path followed to reach this maximum is: {0}".format(flow_dict3))
```

### Analysis & Conclusion 

Since our model keeps track of the residual capacity for the edges, when we increased our constant k from [1,..,2.47], the maximum amount of chocolate fluid transported through the nodes from 1 to 8 kept increasing. But, if we multiply the arc capacities of the edges outflowing from node 1 and those inflowing to node 8 by more than 2.47, the residual capacity of the other edges in the network between node 1 and 8 gets maximized (runs out of capacity) and the edges flips side which signals that that that particular edge would not be able to transfer more fluid (it has been maximized).

Increasing the edges leaving from node 1 and incoming to node 8 only would improve the performance (transfer) to a certain constant. After that constant is achieved, any further increase would not lead to an improvement in our maxflow. The maximum constant k can go to while increasing our maximum flow is 2.47 (Limit).
